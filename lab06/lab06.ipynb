```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
x1 = 0.25
x2 = 0.04 

N = 5424
I0 = 145
R0 = 9
S0 = N - I0 - R0
X0 = [S0, I0, R0]
```

```python
t0 = 0
tmax = 200
dt = 0.01
t = np.arange(t0, tmax, dt)
```

```python
def S1(x, t): #men ili rav
    dx0 = 0
    dx1 = -x2*x[1]
    dx2 = x2*x[1]
    return dx0, dx1, dx2
def S2(x, t): #bolsh
    dx00 = -x1*x[0]
    dx11 = x1*x[0] - x2*x[1]
    dx22 = x2*x[1]
    return dx00, dx11, dx22
```

```python
y1 = odeint(S1, X0, t)
y2 = odeint(S2, X0, t)
```

```python
plt.plot(t, y1[:,0], label='S(t)')
plt.plot(t, y1[:,1], label='I(t)')
plt.plot(t, y1[:,2], label='R(t)')
plt.title('mensh ili ravno')
plt.legend
```

```python
plt.plot(t, y2[:,0], label='S(t)')
plt.plot(t, y2[:,1], label='I(t)')
plt.plot(t, y2[:,2], label='R(t)')
plt.title('bolshe')
plt.legend
```

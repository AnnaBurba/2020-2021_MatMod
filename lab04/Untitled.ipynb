```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
x0 = np.array([1.3, -0.3])

x1 = 18.0
y1 = 0.0

x2 = 2.0
y2 = 8.0

x3 = 7.0
y3 = 3.0
```

```python
def F1(t):
    f = 0
    return f

def F2(t):
    f = 0
    return f

def F3(t):
    f = 3*np.sin(7*t)
    return f
```

```python
t0 = 0
tmax = 73
dt = 0.05
t = np.arange(t0, tmax, dt)
```

```python
def FUNC1(x, t):
    dx1 = x[1]
    dx2 = - x1*x[0] - y1*x[1] - F1(t)
    return dx1, dx2

def FUNC2(x, t):
    dx3 = x[1]
    dx4 = - x2*x[0] - y2*x[1] - F2(t)
    return dx3, dx4

def FUNC3(x, t):
    dx5 = x[1]
    dx6 = - x3*x[0] - y3*x[1] - F3(t)
    return dx5, dx6
```

```python
r1 = odeint(FUNC1, x0, t)
r2 = odeint(FUNC2, x0, t)
r3 = odeint(FUNC3, x0, t)
```

```python
gr1 = r1[:, 0]
gr2 = r1[:, 1]

gr3 = r2[:, 0]
gr4 = r2[:, 1]

gr5 = r3[:, 0]
gr6 = r3[:, 1]
```

```python
plt.plot(gr1, gr2)
```

```python
plt.plot(gr3, gr4)
```
